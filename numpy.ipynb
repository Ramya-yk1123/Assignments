```python
def mean(array):
    n=len(array)
    return np.sum(array)/n
mean(np.array([1,2,3,4,5,6,7]))    
```

```python
def variance(array):
       n=len(array)
       m=mean(array)
       return np.sum((array-m)**2)/n
```

```python
import numpy as np
data=np.array([10,20,30,40,50])
min_val=np.min(data)
max_val=np.max(data)
scaled_data=(np.sum(data)-min_val/max_val-min_val)
print("original data:",data)
print("scaled data:",scaled_data)
variance(np.array([1,2,3,4,5,6]))    
```

```python
import numpy as np
data=np.array([10,20,30,40,50])
m=mean(np.array(data))
v=variance(np.array(data))
s=np.sqrt(v)
standard_scalar=(data-m)/s
print("mean:",m)
print("Variance:",v)
print("deviation:",s)
print("Standard scalar:",standard_scalar)
```
